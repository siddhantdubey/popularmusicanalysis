```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('charts.csv')
data = data.fillna(0)
data.head()
```

```python
data.tail()
```

```python
data.describe()
```

```python
artistdata = data['artist'].value_counts()
print(artistdata.head())
```

```python
artistsongdata.head()
```

```python
print(data[["artist", "song", "peak-rank"]].value_counts().head(10)) #most popular songs of each artist
```

```python
imaginedragonsdata = data[data['artist'] == "Imagine Dragons"]
```

```python
imaginedragonsdata
```

```python
radioactivedata = imaginedragonsdata[imaginedragonsdata['song'] == "Radioactive"]
```

```python
radioactivedata
```

```python
taylorswiftdata = data[data['artist'] == "Taylor Swift"]
```

```python
taylorswiftdata
```

```python
arr = []
fig = plt.figure(figsize=(50,50))
fig.set_facecolor('white')
ax = fig.add_subplot()
legend = []
for song in imaginedragonsdata['song'].unique():
    name_condition = imaginedragonsdata['song']==song
    artist_condition = imaginedragonsdata['artist']=='Imagine Dragons'
    x_value = imaginedragonsdata[name_condition & artist_condition]['date']
    y_value = imaginedragonsdata[name_condition & artist_condition]['rank']
    ax.plot(x_value, y_value)
    legend.append(song)
plt.legend(legend)
plt.show()
```

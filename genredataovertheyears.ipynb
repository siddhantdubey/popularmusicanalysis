```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [20, 20]
```

```python
df = pd.read_csv('yearsandgenres.csv')
```

```python
df.head()
```

```python
df = df.drop(df.columns[0], axis=1)
```

```python
df
```

```python
totals = []
for x in range(1958,2022):
    totals.append(np.sum(df[str(x)]))
```

```python
len(totals)
```

```python
years = [x for x in range(1958, 2022)]
```

```python
plt.plot(years, [(i / j)*100 for i, j in zip(country, totals)], label='Country')
plt.plot(years, [(i / j)*100 for i, j in zip(edm, totals)], label='EDM')
plt.plot(years, [(i / j)*100 for i, j in zip(latin, totals)], label='Latin')
plt.plot(years, [(i / j)*100 for i, j in zip(metal, totals)], label='Metal')
plt.plot(years, [(i / j)*100 for i, j in zip(pop, totals)], label='Pop')
plt.plot(years, [(i / j)*100 for i, j in zip(punk, totals)], label='Punk')
plt.plot(years, [(i / j)*100 for i, j in zip(randb, totals)], label='R&B')
plt.plot(years, [(i / j)*100 for i, j in zip(rap, totals)], label='Rap')
plt.plot(years, [(i / j)*100 for i, j in zip(rockandroll, totals)], label='Rock and Roll')
plt.plot(years, [(i / j)*100 for i, j in zip(reggae, totals)], label='Reggae')
plt.plot(years, [(i / j)*100 for i, j in zip(soul, totals)], label='Soul')
plt.plot(years, [(i / j)*100 for i, j in zip(rock, totals)], label='Rock')
plt.legend()
plt.savefig('genrepopularityovertime.png')
```

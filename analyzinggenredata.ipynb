```python
def map_genre(genres):
    choices = {'Pop':0, 'Rock':0, 'Rock and Roll':0, 'Jazz':0, 'Country':0, 'Rap/Hip-Hop':0, 'R&B':0, 'Soul':0, 'Folk':0, 'EDM':0, 'Latin':0, 'Metal':0, 'Punk':0, 'Ska/Reggae/Danceball':0}
    for genre in genres:
        if 'pop' in genre:
            choices['Pop'] += 1
        elif (('rock-and-roll' in genre) or ('rockabilly' in genre)):
            choices['Rock and Roll'] += 1
        elif 'rock' in genre:
            choices['Rock'] += 1
        elif 'jazz' in genre:
            choices['Jazz'] += 1
        elif 'country' in genre:
            choices['Country'] += 1
        elif (('rap' in genre) or ('hip hop' in genre) ):
            choices['Rap/Hip-Hop'] += 1
        elif ('r&b' in genre):
            choices['R&B'] += 1
        elif ('soul' in genre):
            choices['Soul'] += 1
        elif 'folk' in genre:
            choices['Folk'] += 1
        elif ('edm' in genre) or ('trance' in genre) or ('house' in genre) or ('electro' in genre) or ('future' in genre):
            choices['EDM'] += 1
        elif ('latin' in genre) or ('mexican' in genre):
            choices['Latin'] += 1
        elif ('metal' in genre):
            choices['Metal'] += 1
        elif ('punk' in genre):
            choices['Punk'] += 1
        elif ('reggae' in genre) or ('ska' in genre) or ('danceball' in genre):
            choices['Ska/Reggae/Danceball'] += 1
        else:
            choices['Pop'] += 1
    chosen_genre = max(choices, key=choices.get)
    return chosen_genre
```

```python
cleandf['genre'] = cleandf['genre'].apply(map_genre)
```
